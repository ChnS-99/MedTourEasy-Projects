## Task 5:

```python
# Calculating stock_data returns
stock_returns = stock_data.pct_change()

# Visualizing stock_return
stock_returns.plot();

# Summerizing stock_return
stock_returns.describe()
```

## Task 6:

```python
# calculating benchmark_data return
sp_returns = benchmark_data['S&P 500'].pct_change()

# Visualizing benchmark_data return
sp_returns.plot()

# Summerising benchmark_return
sp_returns.describe()
```

## Task 7:

```python
# Calculating the difference between stock_return and bemchmark_return
excess_returns = stock_returns.sub(sp_returns, axis=0)
 
# Plotting the excess returns
excess_returns.plot()

# Summerising the excess_returns
excess_returns.describe()
```

## Task 8:

```python
# Calculating the average of excess_return
avg_excess_return = excess_returns.mean()

# Visualizing avg_excess_return
avg_excess_return.plot.bar(title='Mean of the excess difference')
```

## Task 9:

```python
# Calculating the standard Deviation
sd_excess_return = excess_returns.std()

# Ploting the standard deviation
sd_excess_return.plot.bar('Standard Deviatin of thr Return Plot')
```

## Task 10:

```python
# Calculating the Sharpe Ratio
daily_sharpe_ratio = avg_excess_return.div (sd_excess_return)

# Annualising the Sharpe Ratio
annual_factor = np.sqrt(252)
annual_sharpe_ratio = daily_sharpe_ratio.mul(annual_factor)

# Visualising Annualised Sharpe Ratio
annual_sharpe_ratio.plot.bar(title = 'Annualized Sharpe Ratio: Stock vs S&P 500')
```
